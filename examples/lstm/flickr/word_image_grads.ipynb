```python
def centered(X):
    return (X - X.min()) / X.max()

def vis_square(in_data, padsize=1, padval=0, num_cols=None):
    data = centered(in_data)
    data = data.transpose(0, 2, 3, 1)
    
    # force the number of filters to be square
    if num_cols is None:
        n = int(np.ceil(np.sqrt(data.shape[0])))
    padding = ((0,
                n ** 2 - data.shape[0]),
                (0, padsize),
                (0, padsize)) + \
              ((0, 0),) * (data.ndim - 3)
    data = np.pad(data, padding, mode='constant', constant_values=(padval, padval))
    
    # tile the filters into an image
    data = data.reshape((n, n) + data.shape[1:]).transpose((0, 2, 1, 3) + tuple(range(4, data.ndim + 1)))
    data = data.reshape((n * data.shape[1], n * data.shape[3]) + data.shape[4:])
    plt.imshow(data)
```

```python
def show_gradients(image_net, image_path, lstm_net, prefix=[], words=['dog', 'water', 'beach']):
    descriptor = image_to_descriptor(image_net, image_path) # run forward before backward
    predict_single_word(lstm_net, descriptor, word_to_ind['<EOS>'])
    for prefix_word in prefix:
        try:
            word = int(prefix_word)
        except ValueError:
            word = word_to_ind[prefix_word]
        predict_single_word(lstm_net, descriptor, word)
    sxs_shape = tuple([len(words) + 1]) + preprocessed_image.shape[1:]
    side_by_side = np.zeros(sxs_shape)
    side_by_side[0] = centered(preprocessed_image[0])
    for index, word in enumerate(words):
        lstm_net.blobs['probs'].diff[:] = 0
        try:
            word = int(word)
        except ValueError:
            word = word_to_ind[word]
        lstm_net.blobs['probs'].diff[0, word, 0, 0] = 1.0
        backward_result = lstm_net.backward()
        image_net.blobs['fc8'].diff[:] = backward_result['image_features']
        image_net_backward_result = image_net.backward()
        side_by_side[index + 1] = centered(image_net_backward_result['data'])
    return side_by_side
```

```python
side_by_side = show_gradients(image_net, image_path, lstm_net, prefix=['a', 'dog'], words=['<EOS>', 'beach', 'water', 'dog', 'water'])
vis_square(side_by_side)
```
